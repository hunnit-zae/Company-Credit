```python
df = df.set_index('Name')
```

```python
df.rename(columns=lambda x: x[:-7], inplace=True)
df
```

```python
#Drop the columns don't use

df_1 = df.drop(columns=['주당배당금', '주당배당금/0'])
```

```python
df_1.isnull().sum()
```

```python
# repalce object value to float

df_1['베타'] = pd.to_numeric(df_1['베타'],errors = 'coerce')
df_1['종업원수'] = pd.to_numeric(df_1['종업원수'],errors = 'coerce')
```

```python
#replace the NaN value to mean

df_1['배당금'] = df_1['배당금'].fillna(df_1['배당금'].mean())
df_1['현금배당금'] = df_1['현금배당금'].fillna(df_1['현금배당금'].mean())
df_1['당좌비율'] = df_1['당좌비율'].fillna(df_1['당좌비율'].mean())
df_1['매출채권회전율'] = df_1['매출채권회전율'].fillna(df_1['매출채권회전율'].mean())
df_1['재고자산회전율1'] = df_1['재고자산회전율1'].fillna(df_1['재고자산회전율1'].mean())
df_1['이익잉여금'] = df_1['이익잉여금'].fillna(df_1['이익잉여금'].mean())
df_1['베타'] = df_1['베타'].fillna(df_1['베타'].mean())
df_1['매출액증가율'] = df_1['매출액증가율'].fillna(df_1['매출액증가율'].mean())
df_1['부채비율'] = df_1['부채비율'].fillna(df_1['부채비율'].mean())
df_1['당기순이익'] = df_1['당기순이익'].fillna(df_1['당기순이익'].mean())
df_1['유동비율'] = df_1['유동비율'].fillna(df_1['유동비율'].mean())
df_1['매출액'] = df_1['매출액'].fillna(df_1['매출액'].mean())
df_1['자기자본증가율'] = df_1['자기자본증가율'].fillna(df_1['자기자본증가율'].mean())
df_1['매출액(종업원1인당)'] = df_1['매출액(종업원1인당)'].fillna(df_1['매출액(종업원1인당)'].mean())
df_1['종업원수'] = df_1['종업원수'].fillna(df_1['종업원수'].mean())
df_1['기말시가총액합계(보통주+우선주)'] = df_1['기말시가총액합계(보통주+우선주)'].fillna(df_1['기말시가총액합계(보통주+우선주)'].mean())
```

```python
# replace the independent variant(Compnay Credit) to binary

df_1['KIS 신용평점'] = df_1['KIS 신용평점'].mask(df_1['KIS 신용평점'] > 6, int(0))
df_1['KIS 신용평점'] = df_1['KIS 신용평점'].mask(df_1['KIS 신용평점'] <= 6, int(1))
```

```python
# replace the value of compnat size 
# small company & not organize = 0
# big company = 1

df_1['기업규모'].mask(df_1['기업규모'] == '중소기업', float(0) , inplace=True)
df_1['기업규모'].mask(df_1['기업규모'] == '기타', float(0) , inplace=True)
df_1['기업규모'].mask(df_1['기업규모'] == '대기업', float(1) , inplace=True)
df_1['기업규모'] = pd.to_numeric(df_1['기업규모'],errors = 'coerce')
```

```python
# sort the 
df_1['1대주주명(지분)'] = df_1['1대주주명(지분)'].str[-6:-1]
df_1['1대주주명(지분)'] = pd.to_numeric(df_1['1대주주명(지분)'],errors = 'coerce')
df_1['1대주주명(지분)'] = df_1['1대주주명(지분)'].fillna(df_1['1대주주명(지분)'].mean())
```
